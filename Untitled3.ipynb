```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.com/s?k=story+books+for+kids&crid=SSGPBY0IDSLE&sprefix=story+books+%2Caps%2C1737&ref=nb_sb_ss_ts-doa-p_2_12"
```

```python
HEADERS = ({'User-Agent':'https://explore.whatismybrowser.com/useragents/parse/?analyse-my-user-agent=yes','Accept-Language': 'en-US, en;q=0.5'})
```

```python
webpage = requests.get(URL,headers = HEADERS)
webpage
```

```python
soup = BeautifulSoup(webpage.content,"html.parser")
soup
```

```python
links = soup.find_all("a",attrs={'class':"a-link-normal s-no-outline"})
links
```

```python
link = links[0].get('href')
link
```

```python
product_list = "https://amazon.com" + link
product_list
```

```python
new_webpage = requests.get(product_list,headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content,"html.parser")
new_soup
```

```python
new_soup.find("span",attrs={"id":'productTitle'}).text.strip()
```

```python
new_soup.find("span",attrs={"class":'a-size-base a-color-price a-color-price'}).text.strip()
```
